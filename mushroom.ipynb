```python
# импорт библиотек
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
# считываем и кодируем данные
data = pd.read_csv('mushrooms.csv')
data['class'] = data['class'].apply(lambda x: ['p', 'e'].index(x))
```

```python
enc = OneHotEncoder()
enc.fit(data[data.columns[1:23]])
```

data['class'] = data['class'].astype(str).astype(float)
data.dtypes

```python
data2 = enc.transform(data[data.columns[1:23]]).toarray()
data2 # получили нампаевский массив
```

```python
# делим данные на тренировочные и тестовые (обучать будем на всех данных)
X = data.loc[:, 'cap-shape':].values
y = data['class'].values
X_train, X_test, y_train, y_test = train_test_split(data2, y) 
print(X_train)
print(y_train)
```

```python
# объявляем модель
gnb = GaussianNB()
gnb.fit(X_train, y_train)
preds = gnb.predict(X_test)
```

```python
metrics.accuracy_score(y_test, preds)
```
